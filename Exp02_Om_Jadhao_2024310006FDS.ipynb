```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets from Seaborn
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
penguins = sns.load_dataset('penguins')

# Titanic Dataset Analysis
print("Titanic Dataset Info:")
print(titanic.info())  # Display information about the Titanic dataset
print(titanic.describe())  # Show basic statistics

# Visualization for Titanic: Count of survivors by class
plt.figure(figsize=(10, 5))
sns.countplot(x='class', hue='survived', data=titanic)
plt.title('Survival Counts by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()  # Show the plot

# Visualization for Titanic: Age distribution of survivors
plt.figure(figsize=(10, 5))
sns.boxplot(x='survived', y='age', data=titanic)
plt.title('Age Distribution by Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()  # Show the plot

# Tips Dataset Analysis
print("\nTips Dataset Info:")
print(tips.info())  # Display information about the Tips dataset
print(tips.describe())  # Show basic statistics

# Visualization for Tips: Scatter plot of tips vs. total bill
plt.figure(figsize=(10, 5))
sns.scatterplot(x='total_bill', y='tip', hue='day', data=tips)
plt.title('Tip vs. Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.legend(title='Day')  # Add legend for days
plt.show()  # Show the plot

# Visualization for Tips: Pie chart of tips by gender
plt.figure(figsize=(8, 8))
gender_counts = tips['sex'].value_counts()  # Count tips by gender
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Tip Proportions by Gender')
plt.axis('equal')  # Equal aspect ratio ensures pie chart is round
plt.show()  # Show the plot

# Penguins Dataset Analysis
print("\nPenguins Dataset Info:")
print(penguins.info())  # Display information about the Penguins dataset
print(penguins.describe())  # Show basic statistics

# Visualization for Penguins: Scatter plot of flipper length vs. body mass
plt.figure(figsize=(10, 6))
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', hue='species', size='body_mass_g', sizes=(20, 200), data=penguins)
plt.title('Bubble Plot: Flipper Length vs. Body Mass')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.legend(title='Species')  # Add legend for species
plt.show()  # Show the plot

# Correlation Heatmap (only for numeric columns)
numeric_penguins = penguins.select_dtypes(include='number')  # Select only numeric columns
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_penguins.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()  # Show the plot
```
